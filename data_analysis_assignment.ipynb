# **Part A: Theoretical Fundamentals (Maximum Likelihood & Decision Space)**

**derivation of why the raw Maximum Likelihood Estimator for the covariance matrix,
 ̂Σ MLE is inherently biased in finite samples**

 Let $\mathbf{X}_1, \mathbf{X}_2, \dots, \mathbf{X}_n$ be an independent and identically distributed ($i.i.d.$) random sample from a multivariate population with mean vector $\boldsymbol{\mu} \in \mathbb{R}^p$ and covariance matrix $\boldsymbol{\Sigma} \in \mathbb{R}^{p \times p}$.The sample mean vector is defined as:$$\overline{\mathbf{X}} = \frac{1}{n} \sum_{i=1}^n \mathbf{X}_i$$The raw Maximum Likelihood Estimator (MLE) for the covariance matrix, denoted here as $\widehat{\boldsymbol{\Sigma}}_{\text{raw}}$, is:$$\widehat{\boldsymbol{\Sigma}}_{\text{raw}} = \frac{1}{n} \sum_{i=1}^n (\mathbf{X}_i - \overline{\mathbf{X}})(\mathbf{X}_i - \overline{\mathbf{X}})^T$$To evaluate the expectation $\mathbb{E}[\widehat{\boldsymbol{\Sigma}}_{\text{raw}}]$, we express the deviation from the sample mean using the true population mean $\boldsymbol{\mu}$:$$\mathbf{X}_i - \overline{\mathbf{X}} = (\mathbf{X}_i - \boldsymbol{\mu}) - (\overline{\mathbf{X}} - \boldsymbol{\mu})$$Expanding the outer product yields:$$\sum_{i=1}^n (\mathbf{X}_i - \overline{\mathbf{X}})(\mathbf{X}_i - \overline{\mathbf{X}})^T = \sum_{i=1}^n \left[ (\mathbf{X}_i - \boldsymbol{\mu}) - (\overline{\mathbf{X}} - \boldsymbol{\mu}) \right] \left[ (\mathbf{X}_i - \boldsymbol{\mu}) - (\overline{\mathbf{X}} - \boldsymbol{\mu}) \right]^T$$$$= \sum_{i=1}^n (\mathbf{X}_i - \boldsymbol{\mu})(\mathbf{X}_i - \boldsymbol{\mu})^T - \sum_{i=1}^n (\mathbf{X}_i - \boldsymbol{\mu})(\overline{\mathbf{X}} - \boldsymbol{\mu})^T - \sum_{i=1}^n (\overline{\mathbf{X}} - \boldsymbol{\mu})(\mathbf{X}_i - \boldsymbol{\mu})^T + \sum_{i=1}^n (\overline{\mathbf{X}} - \boldsymbol{\mu})(\overline{\mathbf{X}} - \boldsymbol{\mu})^T$$Since $\sum_{i=1}^n (\mathbf{X}_i - \boldsymbol{\mu}) = n(\overline{\mathbf{X}} - \boldsymbol{\mu})$, the two middle cross-product terms compress:$$\sum_{i=1}^n (\mathbf{X}_i - \overline{\mathbf{X}})(\mathbf{X}_i - \overline{\mathbf{X}})^T = \sum_{i=1}^n (\mathbf{X}_i - \boldsymbol{\mu})(\mathbf{X}_i - \boldsymbol{\mu})^T - n(\overline{\mathbf{X}} - \boldsymbol{\mu})(\overline{\mathbf{X}} - \boldsymbol{\mu})^T$$Taking expectations on both sides:$$\mathbb{E}\left[ \sum_{i=1}^n (\mathbf{X}_i - \overline{\mathbf{X}})(\mathbf{X}_i - \overline{\mathbf{X}})^T \right] = \sum_{i=1}^n \mathbb{E}\left[ (\mathbf{X}_i - \boldsymbol{\mu})(\mathbf{X}_i - \boldsymbol{\mu})^T \right] - n \mathbb{E}\left[ (\overline{\mathbf{X}} - \boldsymbol{\mu})(\overline{\mathbf{X}} - \boldsymbol{\mu})^T \right]$$By definition of the population covariance matrix:$$\mathbb{E}\left[ (\mathbf{X}_i - \boldsymbol{\mu})(\mathbf{X}_i - \boldsymbol{\mu})^T \right] = \boldsymbol{\Sigma}$$For the sample mean vector, its variance-covariance matrix is given by:$$\mathbb{E}\left[ (\overline{\mathbf{X}} - \boldsymbol{\mu})(\overline{\mathbf{X}} - \boldsymbol{\mu})^T \right] = \text{Var}(\overline{\mathbf{X}}) = \frac{1}{n}\boldsymbol{\Sigma}$$Substituting these expectations back into our sum equation:$$\mathbb{E}\left[ \sum_{i=1}^n (\mathbf{X}_i - \overline{\mathbf{X}})(\mathbf{X}_i - \overline{\mathbf{X}})^T \right] = \sum_{i=1}^n \boldsymbol{\Sigma} - n \left( \frac{1}{n}\boldsymbol{\Sigma} \right) = n\boldsymbol{\Sigma} - \boldsymbol{\Sigma} = (n - 1)\boldsymbol{\Sigma}$$Multiplying by the scalar $1/n$ from the raw MLE definition:$$\mathbb{E}[\widehat{\boldsymbol{\Sigma}}_{\text{raw}}] = \frac{1}{n} (n - 1)\boldsymbol{\Sigma} = \left(1 - \frac{1}{n}\right)\boldsymbol{\Sigma}$$This proves that $\widehat{\boldsymbol{\Sigma}}_{\text{raw}}$ is inherently biased downward in finite samples, underestimating the true population covariance structure by a factor of $(n-1)/n$.



 **2.Mathematical Application of Bessel's CorrectionTo isolate an unbiased unique estimator $\mathbf{S}$**

  we require $\mathbb{E}[\mathbf{S}] = \boldsymbol{\Sigma}$. Given that:$$\mathbb{E}\left[ \sum_{i=1}^n (\mathbf{X}_i - \overline{\mathbf{X}})(\mathbf{X}_i - \overline{\mathbf{X}})^T \right] = (n - 1)\boldsymbol{\Sigma}$$We scale both sides by the reciprocal scalar factor $\frac{1}{n - 1}$ (Bessel’s Correction):$$\mathbb{E}\left[ \frac{1}{n - 1} \sum_{i=1}^n (\mathbf{X}_i - \overline{\mathbf{X}})(\mathbf{X}_i - \overline{\mathbf{X}})^T \right] = \boldsymbol{\Sigma}$$Thus, the sample covariance matrix $\mathbf{S}$ defined below constitutes the unique, unbiased estimator:$$\mathbf{S} = \frac{1}{n - 1} \sum_{i=1}^n (\mathbf{X}_i - \overline{\mathbf{X}})(\mathbf{X}_i - \overline{\mathbf{X}})^T$$
  
  **3. Error Definitions within Continuous Structural Health Monitoring (SHM)**
  
  Type I Error ($\alpha$): The probability of committing a false alarm. In SHM, this occurs when the diagnostic layer rejects the null hypothesis $H_0$ (Asset is Healthy) when the asset is, in fact, operating within normal baseline conditions. This results in costly, unnecessary downtime, false maintenance alerts, and operational disruptions.
  
  Type II Error ($\beta$): The probability of a missed detection. In SHM, this occurs when the diagnostic layer fails to reject the null hypothesis $H_0$ despite the structure suffering an active anomaly, mechanical drift, or physical fracture. This leads to catastrophic failure because unmitigated degradation goes unnoticed.
  
  **4. Consequence of Ultra-Conservative Significance ThresholdsSetting an ultra-conservative alpha threshold**
  
   (e.g., $\alpha = 10^{-6}$) shifts the critical boundary of the diagnostic test statistic outward. Because the decision boundary is shared between the null distribution $H_0$ and the alternative distribution $H_1$, minimizing $\alpha$ causes the critical region to contract dramatically.The mathematical consequences are as follows:Statistical Power Degradation: Since $\alpha \to 0$, the probability of a Type II error ($\beta$) increases significantly because the detection threshold is high. Consequently, the statistical power ($1 - \beta$) drops, making the system blind to subtle or early-stage structural faults.Geometric Volume Expansion of the Confidence Ellipsoid: The healthy operational boundary is modeled as a Hotelling's $T^2$ confidence ellipsoid. The volume of this hyper-ellipsoid is inversely proportional to the chosen significance level $\alpha$. Specifically, as $\alpha \to 0$, the critical value $T^2_{\text{crit}}$ expands outward to capture $1 - \alpha$ ($99.9999\%$) of the probability mass. Geometrically, this inflates the volume of the "healthy operation" zone, absorbing true structural drifts and low-magnitude damages inside its expanded boundary, which prevents them from triggering an alarm.

# **Part B: Theoretical Extension (Asymptotic Distributions & Slutsky's Theorem)**

**1. Rigorous Statement of Slutsky's Theorem**

Let $\{\mathbf{Y}_n\}$ and $\{\mathbf{Z}_n\}$ be sequences of multivariate random vectors and matrices, respectively. If $\mathbf{Y}_n$ converges in distribution to a random variable $\mathbf{Y}$ (denoted $\mathbf{Y}_n \xrightarrow{d} \mathbf{Y}$), and $\mathbf{Z}_n$ converges in probability to a deterministic constant matrix $\mathbf{C}$ (denoted $\mathbf{Z}_n \xrightarrow{p} \mathbf{C}$), then:$$\mathbf{Z}_n \mathbf{Y}_n \xrightarrow{d} \mathbf{C}\mathbf{Y}$$$$\mathbf{Z}_n^{-1} \mathbf{Y}_n \xrightarrow{d} \mathbf{C}^{-1}\mathbf{Y} \quad (\text{provided } \mathbf{C} \text{ is non-singular})$$

**2. Justification of Empirical Covariance Substitution via Slutsky's TheoremBy the Multivariate Central Limit Theorem (MCLT)**

The standardized sample mean vector satisfies:$$\mathbf{Y}_n = \sqrt{n}(\overline{\mathbf{X}}_n - \boldsymbol{\mu}) \xrightarrow{d} \mathcal{N}_p(\mathbf{0}, \boldsymbol{\Sigma})$$By the Weak Law of Large Numbers (WLLN), the unbiased empirical sample covariance matrix $\mathbf{S}_n$ is a consistent estimator of the latent matrix $\boldsymbol{\Sigma}$:$$\mathbf{S}_n \xrightarrow{p} \boldsymbol{\Sigma} \quad \text{as } n \to \infty$$Because matrix inversion and matrix square roots are continuous mappings on the space of positive-definite matrices, continuous mapping principles dictate that:$$\mathbf{S}_n^{-1/2} \xrightarrow{p} \boldsymbol{\Sigma}^{-1/2} \quad \text{as } n \to \infty$$We now apply Slutsky's Theorem by setting our sequence of random vectors as $\mathbf{Y}_n = \sqrt{n}(\overline{\mathbf{X}}_n - \boldsymbol{\mu})$ and our sequence of scaling matrices as $\mathbf{Z}_n = \mathbf{S}_n^{-1/2}$. Since $\mathbf{Y}_n \xrightarrow{d} \mathcal{N}_p(\mathbf{0}, \boldsymbol{\Sigma})$ and $\mathbf{Z}_n \xrightarrow{p} \boldsymbol{\Sigma}^{-1/2}$:$$\mathbf{S}_n^{-1/2} \left[ \sqrt{n}(\overline{\mathbf{X}}_n - \boldsymbol{\mu}) \right] \xrightarrow{d} \boldsymbol{\Sigma}^{-1/2} \mathbf{X}_Z \quad \text{where } \mathbf{X}_Z \sim \mathcal{N}_p(\mathbf{0}, \boldsymbol{\Sigma})$$The linear transformation of a Gaussian distribution preserves its normality while transforming its covariance. The resulting asymptotic covariance is:$$\text{Cov}(\boldsymbol{\Sigma}^{-1/2} \mathbf{X}_Z) = \boldsymbol{\Sigma}^{-1/2} \boldsymbol{\Sigma} \boldsymbol{\Sigma}^{-1/2} = \mathbf{I}_p$$Therefore:$$\sqrt{n}\mathbf{S}_n^{-1/2}(\overline{\mathbf{X}}_n - \boldsymbol{\mu}) \xrightarrow{d} \mathcal{N}_p(\mathbf{0}, \mathbf{I}_p)$$This mathematically justifies substituting the observable sample metric $\mathbf{S}_n$ for the unobservable latent matrix $\boldsymbol{\Sigma}$ in automated diagnostic dashboards.

Given that our calculated $p$-value ($\approx 3.98 \times 10^{-12}$) falls far below the designated significance threshold ($\alpha = 0.01$), we reject the null hypothesis $H_0$ of first-moment homogeneity.Operational Verdict: The baseline center has shifted significantly over time.
```python
import numpy as np
import pandas as pd
import scipy.stats as stats

# --- DO NOT MODIFY: Synthetic Strain Sensor Data Generator ---
np.random.seed(42)
n_samples = 5000
n_features = 3

# Generating base stationary multivariate normal noise
base_data = np.random.multivariate_normal(
    mean=[0.5, -0.2, 1.1],
    cov=[[0.09, 0.02, 0.01],
         [0.02, 0.06, 0.03],
         [0.01, 0.03, 0.05]],
    size=n_samples)

# Injecting a subtle, hidden structural drift in the final 1,000 snapshots
base_data[4000:, 0] += 0.015  # Sensor 1 drift
base_data[4000:, 2] -= 0.010  # Sensor 3 drift

df_strain = pd.DataFrame(base_data, columns=['Strain_Ch1', 'Strain_Ch2', 'Strain_Ch3'])
# ----------------------------------------------------------------

def verify_first_moment_homogeneity(df: pd.DataFrame, g_chunks: int = 5) -> dict:
    """
    Partitions the dataset into g chunks and evaluates first-moment homogeneity
    via Wilks' Lambda and Bartlett's Chi-Square asymptotic transformation.
    """
    X = df.to_numpy()
    N, p = X.shape
    n_k = N // g_chunks  # Size of each chunk

    # 1. Compute global mean
    global_mean = np.mean(X, axis=0)

    # Initialize Within (W) and Between (B) matrices
    W = np.zeros((p, p))
    B = np.zeros((p, p))

    # 2. Iterate through groups to accumulate matrices
    for i in range(g_chunks):
        chunk = X[i * n_k : (i + 1) * n_k, :]
        chunk_mean = np.mean(chunk, axis=0)

        # Within-chunk scatter contribution
        dev_within = chunk - chunk_mean
        W += dev_within.T @ dev_within

        # Between-chunk scatter contribution
        dev_between = chunk_mean - global_mean
        B += n_k * np.outer(dev_between, dev_between)

    # 3. Calculate Wilks' Lambda (det(W) / det(W + B))
    det_W = np.linalg.det(W)
    det_Total = np.linalg.det(W + B)
    wilks_lambda = det_W / det_Total

    # 4. Apply Bartlett's Chi-Square Asymptotic Transformation
    # Formula factor: - (N - 1 - (p + g_chunks) / 2) * ln(Wilks' Lambda)
    factor = N - 1.0 - (p + g_chunks) / 2.0
    bartlett_stat = -factor * np.log(wilks_lambda)

    # Degrees of freedom: p * (g - 1)
    df_chi = p * (g_chunks - 1)
    p_value = 1.0 - stats.chi2.cdf(bartlett_stat, df_chi)

    return {
        "Wilks_Lambda": wilks_lambda,
        "Bartlett_Stat": bartlett_stat,
        "p_value": p_value,
        "Degrees_of_Freedom": df_chi
    }

# Execute completed validation layer
results = verify_first_moment_homogeneity(df_strain, g_chunks=5)
print("--- FIRST-MOMENT HOMOGENEITY DIAGNOSTICS ---")
for key, val in results.items():
    print(f"{key}: {val:.6f}" if isinstance(val, float) else f"{key}: {val}")
```

# **SECTION 2: Geometric Subspace Optimization via Principal Component Analysis (PCA)**

## **Part A: Theoretical Fundamentals (Coordinate Projections & Orthogonality)**

**1. Proof of Strict Covariance Diagonalization**

We are given a zero-mean centered observation vector $\mathbf{x} \in \mathbb{R}^p$ with population covariance matrix $\boldsymbol{\Sigma} = \mathbb{E}[\mathbf{x}\mathbf{x}^T]$. The spectral decomposition of $\boldsymbol{\Sigma}$ is:$$\boldsymbol{\Sigma} = \mathbf{P} \boldsymbol{\Lambda} \mathbf{P}^T$$where $\mathbf{P}$ is an orthogonal matrix ($\mathbf{P}^T\mathbf{P} = \mathbf{P}\mathbf{P}^T = \mathbf{I}$) containing eigenvectors, and $\boldsymbol{\Lambda}$ is a diagonal matrix containing eigenvalues $\lambda_1 \ge \lambda_2 \ge \dots \ge \lambda_p$.The transformed principal coordinate mapping is defined as:$$\mathbf{y} = \mathbf{P}^T \mathbf{x}$$The theoretical covariance matrix of the transformed vector $\mathbf{y}$, denoted $\boldsymbol{\Sigma}_y$, is calculated as follows:$$\boldsymbol{\Sigma}_y = \mathbb{E}[\mathbf{y}\mathbf{y}^T] = \mathbb{E}[(\mathbf{P}^T\mathbf{x})(\mathbf{P}^T\mathbf{x})^T]$$Using the linear algebra identity $(\mathbf{A}\mathbf{B})^T = \mathbf{B}^T\mathbf{A}^T$:$$\boldsymbol{\Sigma}_y = \mathbb{E}[\mathbf{P}^T \mathbf{x} \mathbf{x}^T \mathbf{P}]$$Since the matrix $\mathbf{P}$ is a deterministic transformation matrix derived from the population parameter, it passes outside the expectation operator:$$\boldsymbol{\Sigma}_y = \mathbf{P}^T \mathbb{E}[\mathbf{x} \mathbf{x}^T] \mathbf{P} = \mathbf{P}^T \boldsymbol{\Sigma} \mathbf{P}$$Substituting the spectral decomposition $\boldsymbol{\Sigma} = \mathbf{P} \boldsymbol{\Lambda} \mathbf{P}^T$:$$\boldsymbol{\Sigma}_y = \mathbf{P}^T (\mathbf{P} \boldsymbol{\Lambda} \mathbf{P}^T) \mathbf{P}$$Regrouping using matrix associativity:$$\boldsymbol{\Sigma}_y = (\mathbf{P}^T \mathbf{P}) \boldsymbol{\Lambda} (\mathbf{P}^T \mathbf{P})$$Because $\mathbf{P}$ is orthogonal, $\mathbf{P}^T \mathbf{P} = \mathbf{I}$. Therefore:$$\boldsymbol{\Sigma}_y = \mathbf{I} \boldsymbol{\Lambda} \mathbf{I} = \boldsymbol{\Lambda}$$

This proves that the transformed coordinates have a strictly diagonal covariance matrix $\boldsymbol{\Lambda}$.Geometric & Statistical InterpretationStatistically, a diagonal covariance matrix means that the cross-covariance between any two distinct principal coordinates $y_j$ and $y_k$ ($j \neq k$) is exactly zero ($\text{Cov}(y_j, y_k) = 0$). Geometrically, this means that the original coupled data axes have been rotated into a new orthogonal coordinate system aligned with the principal axes of the data's concentration ellipsoid. This eliminates all linear redundancy and cross-talk between sensors.

**2. Proof of Total Variance Conservation via Trace PermutationThe total variance of a multivariate system in its raw space is the sum of the variances of individual variables, which equals the trace of its covariance matrix $\text{Tr}(\boldsymbol{\Sigma})$.**

Using our previous relation $\boldsymbol{\Sigma}_y = \boldsymbol{\Lambda} = \mathbf{P}^T \boldsymbol{\Sigma} \mathbf{P}$, we can isolate $\boldsymbol{\Sigma}$ by pre-multiplying by $\mathbf{P}$ and post-multiplying by $\mathbf{P}^T$:$$\boldsymbol{\Sigma} = \mathbf{P} \boldsymbol{\Lambda} \mathbf{P}^T$$Taking the trace of both sides:$$\text{Tr}(\boldsymbol{\Sigma}) = \text{Tr}(\mathbf{P} \boldsymbol{\Lambda} \mathbf{P}^T)$$The trace cyclic permutation property states that $\text{Tr}(\mathbf{A}\mathbf{B}\mathbf{C}) = \text{Tr}(\mathbf{C}\mathbf{A}\mathbf{B})$. Setting $\mathbf{A} = \mathbf{P}$, $\mathbf{B} = \boldsymbol{\Lambda}$, and $\mathbf{C} = \mathbf{P}^T$:$$\text{Tr}(\boldsymbol{\Sigma}) = \text{Tr}(\mathbf{P}^T \mathbf{P} \boldsymbol{\Lambda})$$Since $\mathbf{P}^T \mathbf{P} = \mathbf{I}$:$$\text{Tr}(\boldsymbol{\Sigma}) = \text{Tr}(\mathbf{I} \boldsymbol{\Lambda}) = \text{Tr}(\boldsymbol{\Lambda})$$Because $\boldsymbol{\Lambda}$ is the covariance matrix of our principal components $\mathbf{y}$, $\text{Tr}(\boldsymbol{\Lambda}) = \text{Tr}(\boldsymbol{\Sigma}_y)$. Thus:$$\sum_{j=1}^p \text{Var}(x_j) = \sum_{j=1}^p \lambda_j = \sum_{j=1}^p \text{Var}(y_j)$$This completes the proof that total system variance is preserved under the orthogonal projection mapping.

**3. Mathematical Subspace Partition FormulationsWhen truncating to the first $k$ principal components:Cumulative Explained Variance Ratio ($\psi_k$):**

$$\psi_k = \frac{\sum_{j=1}^k \lambda_j}{\text{Tr}(\boldsymbol{\Sigma})} = \frac{\sum_{j=1}^k \lambda_j}{\sum_{m=1}^p \lambda_m}$$Residual Unexplained Variance Ratio ($\zeta_k$):$$\zeta_k = \frac{\sum_{j=k+1}^p \lambda_j}{\text{Tr}(\boldsymbol{\Sigma})} = \frac{\sum_{j=k+1}^p \lambda_j}{\sum_{m=1}^p \lambda_m} = 1 - \psi_k$$

**4. Verification of Orthogonal Subspace Energy Decomposition (Pythagorean Identity)**

Let the projection matrix onto the retained $k$-dimensional principal subspace be $\mathbf{P}_k \in \mathbb{R}^{p \times k}$. The reconstructed observation vector is:$$\hat{\mathbf{x}} = \mathbf{P}_k \mathbf{P}_k^T \mathbf{x}$$The residual error vector is defined as:$$\mathbf{e} = \mathbf{x} - \hat{\mathbf{x}} = (\mathbf{I} - \mathbf{P}_k \mathbf{P}_k^T)\mathbf{x}$$To evaluate the squared Euclidean norm $\|\mathbf{x}\|^2 = \mathbf{x}^T \mathbf{x}$:$$\mathbf{x}^T \mathbf{x} = (\hat{\mathbf{x}} + \mathbf{e})^T (\hat{\mathbf{x}} + \mathbf{e}) = \hat{\mathbf{x}}^T \hat{\mathbf{x}} + \hat{\mathbf{x}}^T \mathbf{e} + \mathbf{e}^T \hat{\mathbf{x}} + \mathbf{e}^T \mathbf{e}$$Let us analyze the inner product cross term $\hat{\mathbf{x}}^T \mathbf{e}$:$$\hat{\mathbf{x}}^T \mathbf{e} = (\mathbf{P}_k \mathbf{P}_k^T \mathbf{x})^T \left( (\mathbf{I} - \mathbf{P}_k \mathbf{P}_k^T)\mathbf{x} \right) = \mathbf{x}^T \mathbf{P}_k \mathbf{P}_k^T (\mathbf{I} - \mathbf{P}_k \mathbf{P}_k^T)\mathbf{x}$$$$= \mathbf{x}^T (\mathbf{P}_k \mathbf{P}_k^T - \mathbf{P}_k \mathbf{P}_k^T \mathbf{P}_k \mathbf{P}_k^T)\mathbf{x}$$Because the projection matrices are orthonormal, $\mathbf{P}_k^T \mathbf{P}_k = \mathbf{I}_k$. Thus:$$\mathbf{P}_k \mathbf{P}_k^T \mathbf{P}_k \mathbf{P}_k^T = \mathbf{P}_k (\mathbf{I}_k) \mathbf{P}_k^T = \mathbf{P}_k \mathbf{P}_k^T$$Substituting this back reveals that the core expression cancels out:$$\hat{\mathbf{x}}^T \mathbf{e} = \mathbf{x}^T (\mathbf{P}_k \mathbf{P}_k^T - \mathbf{P}_k \mathbf{P}_k^T)\mathbf{x} = 0$$Since the scalar cross terms are zero ($\hat{\mathbf{x}}^T \mathbf{e} = \mathbf{e}^T \mathbf{h} = 0$), the equation simplifies to:$$\mathbf{x}^T \mathbf{x} = \hat{\mathbf{x}}^T \hat{\mathbf{x}} + \mathbf{e}^T \mathbf{e}$$$$\|\mathbf{x}\|^2 = \|\hat{\mathbf{x}}\|^2 + \|\mathbf{e}\|^2$$This completes the proof that total energy splits cleanly into independent subspace energies.

**5. Diagnostic Functions and Anomaly MappingHotelling's $T^2$ Statistic:**

 Measures Mahalanobis distance within the bounded, principal hyper-ellipsoid subspace. It tracks changes in the magnitude of normal operational modes.Squared Prediction Error (SPE) / $Q$ Statistic: Measures the orthogonal squared Euclidean distance from the principal subspace. It tracks deviations that break the structural correlation rules established during baseline calibration.Matching Asset Event ScenariosEnvironmental Load Anomaly (e.g., Extreme Wind): Matches with Hotelling's $T^2$. High wind loads do not alter the underlying physical structure or load paths of the plant; they simply push the structural responses further along normal operational profiles. This causes an inflation of variance within the known principal components, making $T^2$ highly sensitive while SPE remains stable.Internal Structural Fracture (e.g., Fatigue Crack): Matches with SPE ($Q$). A fatigue crack breaks the physical connection between components, altering the structural system's load path. This introduces a new direction of variance that is orthogonal to the baseline PCA model, causing the residual error vector $\mathbf{e}$ to spike and triggering a sharp increase in the SPE statistic.

# **Part B: Numerical Verification**

```python
import numpy as np

def compute_pca_subspace_metrics(X_raw: np.ndarray) -> dict:
    """
    Standardizes input data, performs spectral decomposition on the sample covariance,
    and maps the expected value profiles for Hotelling's T^2 and SPE across all truncation thresholds.
    """
    N, p = X_raw.shape
    # 1. Standardize and center realizations
    X_mean = np.mean(X_raw, axis=0)
    X_std = np.std(X_raw, axis=0, ddof=1)
    X_scaled = (X_raw - X_mean) / X_std

    # 2. Compute unbiased sample covariance matrix
    S = (X_scaled.T @ X_scaled) / (N - 1)

    # Eigen-decomposition
    eigenvalues, eigenvectors = np.linalg.eigh(S)

    # Sort in descending order
    idx = np.argsort(eigenvalues)[::-1]
    lambdas = eigenvalues[idx]
    P = eigenvectors[:, idx]

    t2_means = []
    spe_means = []
    cutoffs = list(range(1, p + 1))

    # 3. Evaluate every possible subspace cutoff
    for k in cutoffs:
        P_k = P[:, :k]  # Retained loading space
        L_k_inv = np.diag(1.0 / lambdas[:k])  # Inverse eigenvalue weights

        # Project scores for T^2 calculation
        scores_k = X_scaled @ P_k  # Shape: (N, k)
        t2_all = np.sum((scores_k @ L_k_inv) * scores_k, axis=1)

        # Reconstruction error for SPE
        recon = scores_k @ P_k.T
        residuals = X_scaled - recon
        spe_all = np.sum(residuals**2, axis=1)

        t2_means.append(np.mean(t2_all))
        spe_means.append(np.mean(spe_all))

    return {
        "Cutoffs": cutoffs,
        "Eigenvalues": lambdas.tolist(),
        "Mean_T2": t2_means,
        "Mean_SPE": spe_means
    }

# Synthetic verification utilizing df_strain dimension structure
X_verify = df_strain.to_numpy()
pca_results = compute_pca_subspace_metrics(X_verify)

print("--- GEOMETRIC SUBSPACE DIAGNOSTICS ---")
print(f"Sorted Eigenvalues: {np.round(pca_results['Eigenvalues'], 5)}")
print(f"Mean Hotelling T^2 per cutoff k: {np.round(pca_results['Mean_T2'], 5)}")
print(f"Mean SPE (Q) per cutoff k:       {np.round(pca_results['Mean_SPE'], 5)}")
```

# **SECTION 3: Latent Subspace Decomposition via Factor Analysis (FA)**

## **Part A: Theoretical Fundamentals (Generative Model & Commonalities)**

**1. Proof of the Fundamental Equation of Factor Analysis**

We are given the normalized vector $\mathbf{z} \in \mathbb{R}^p$ modeled as:$$\mathbf{z} = \boldsymbol{\Lambda} \mathbf{f} + \boldsymbol{\epsilon}$$where $\mathbf{f} \in \mathbb{R}^m$ represents the latent factors, and $\boldsymbol{\epsilon} \in \mathbb{R}^p$ represents the unique sensor noises. The model assumptions are:$$\mathbb{E}[\mathbf{f}] = \mathbf{0}, \quad \text{Var}(\mathbf{f}) = \mathbb{E}[\mathbf{f}\mathbf{f}^T] = \mathbf{I}_m$$$$\mathbb{E}[\boldsymbol{\epsilon}] = \mathbf{0}, \quad \text{Var}(\boldsymbol{\epsilon}) = \mathbb{E}[\boldsymbol{\epsilon}\boldsymbol{\epsilon}^T] = \boldsymbol{\Psi} = \text{diag}(\psi_1^2, \dots, \psi_p^2)$$$$\text{Cov}(\mathbf{f}, \boldsymbol{\epsilon}) = \mathbb{E}[\mathbf{f}\boldsymbol{\epsilon}^T] = \mathbf{0}_{m \times p}$$The population correlation matrix is $\boldsymbol{\Sigma} = \mathbb{E}[\mathbf{z}\mathbf{z}^T]$ (since $\mathbf{z}$ is standardized). Substituting the model equation into the covariance definition:$$\boldsymbol{\Sigma} = \mathbb{E}[(\boldsymbol{\Lambda}\mathbf{f} + \boldsymbol{\epsilon})(\boldsymbol{\Lambda}\mathbf{f} + \boldsymbol{\epsilon})^T]$$Expanding using linear algebra transposition properties:$$\boldsymbol{\Sigma} = \mathbb{E}\left[ (\boldsymbol{\Lambda}\mathbf{f} + \boldsymbol{\epsilon})(\mathbf{f}^T\boldsymbol{\Lambda}^T + \boldsymbol{\epsilon}^T) \right]$$$$\boldsymbol{\Sigma} = \mathbb{E}[\boldsymbol{\Lambda}\mathbf{f}\mathbf{f}^T\boldsymbol{\Lambda}^T] + \mathbb{E}[\boldsymbol{\Lambda}\mathbf{f}\boldsymbol{\epsilon}^T] + \mathbb{E}[\boldsymbol{\epsilon}\mathbf{f}^T\boldsymbol{\Lambda}^T] + \mathbb{E}[\boldsymbol{\epsilon}\boldsymbol{\epsilon}^T]$$By linearity of expectation, we pull out the deterministic loading matrices:$$\boldsymbol{\Sigma} = \boldsymbol{\Lambda}\mathbb{E}[\mathbf{f}\mathbf{f}^T]\boldsymbol{\Lambda}^T + \boldsymbol{\Lambda}\mathbb{E}[\mathbf{f}\boldsymbol{\epsilon}^T] + \mathbb{E}[\boldsymbol{\epsilon}\mathbf{f}^T]\boldsymbol{\Lambda}^T + \mathbb{E}[\boldsymbol{\epsilon}\boldsymbol{\epsilon}^T]$$Substituting the model assumptions ($\mathbb{E}[\mathbf{f}\mathbf{f}^T] = \mathbf{I}$, $\mathbb{E}[\mathbf{f}\boldsymbol{\epsilon}^T] = \mathbf{0}$, $\mathbb{E}[\boldsymbol{\epsilon}\boldsymbol{\epsilon}^T] = \boldsymbol{\Psi}$):$$\boldsymbol{\Sigma} = \boldsymbol{\Lambda}(\mathbf{I}_m)\boldsymbol{\Lambda}^T + \boldsymbol{\Lambda}(\mathbf{0}) + (\mathbf{0})\boldsymbol{\Lambda}^T + \boldsymbol{\Psi}$$$$\boldsymbol{\Sigma} = \boldsymbol{\Lambda}\boldsymbol{\Lambda}^T + \boldsymbol{\Psi}$$This completes the derivation of the Fundamental Equation of Factor Analysis.Parameter Interpretation ($i$-th Diagonal Entry)For a specific sensor channel $i$, the total variance expands as:$$\text{Var}(z_i) = 1 = \sum_{j=1}^m \lambda_{ij}^2 + \psi_i^2$$Communality ($h_i^2 = \sum_{j=1}^m \lambda_{ij}^2$): The portion of a sensor's variance shared with other sensors through the latent structural modes. In asset diagnostics, a high communality means the sensor is highly integrated into the system-wide structural behavior.Uniqueness ($\psi_i^2$): The portion of variance specific to that sensor, representing localized measurement noise or instrumentation errors. In asset diagnostics, a sudden increase in uniqueness identifies a localized sensor fault rather than global structural damage.

**2. Orthogonal Matrix Rotations and InvarianceRaw PCA vs. Varimax-Rotated FA**

 Loading Vector: A raw PCA loading vector points in the direction of maximum total variance, which often results in dense loading vectors where every sensor has non-zero weights across all components. A Varimax-rotated factor loading matrix applies an orthogonal rotation to maximize the variance of the squared loadings within each column. This creates a "simple structure" where each sensor loads heavily onto one factor and near zero onto others, mapping sensors directly to independent physical processes.Mathematical Invariance of Communality: Let $\mathbf{T}$ be an orthogonal rotation matrix satisfying $\mathbf{T}\mathbf{T}^T = \mathbf{T}^T\mathbf{T} = \mathbf{I}$. The rotated loading matrix is $\boldsymbol{\Lambda}^* = \boldsymbol{\Lambda}\mathbf{T}$. Substituting this into the variance factorization:$$\boldsymbol{\Lambda}^*(\boldsymbol{\Lambda}^*)^T = (\boldsymbol{\Lambda}\mathbf{T})(\boldsymbol{\Lambda}\mathbf{T})^T = \boldsymbol{\Lambda}\mathbf{T}\mathbf{T}^T\boldsymbol{\Lambda}^T = \boldsymbol{\Lambda}(\mathbf{I})\boldsymbol{\Lambda}^T = \boldsymbol{\Lambda}\boldsymbol{\Lambda}^T$$Because the product $\boldsymbol{\Lambda}\boldsymbol{\Lambda}^T$ remains completely unchanged, its diagonal elements ($h_i^2$) are invariant under orthogonal rotations. This means rotation redistributes variance across the latent factors to improve interpretability without altering the total explained variance or individual sensor noise floors.

 **3. Derivation of Thomson's Regression Score Matrix Equation**

 We construct a joint random vector containing the observed data $\mathbf{z}$ and the unobservable latent factors $\mathbf{f}$. Since $\mathbb{E}[\mathbf{z}]=\mathbf{0}$ and $\mathbb{E}[\mathbf{f}]=\mathbf{0}$, the joint covariance matrix is:$$\mathbf{Cov} \begin{pmatrix} \mathbf{z} \\ \mathbf{f} \end{pmatrix} = \begin{pmatrix} \mathbb{E}[\mathbf{z}\mathbf{z}^T] & \mathbb{E}[\mathbf{z}\mathbf{f}^T] \\ \mathbb{E}[\mathbf{f}\mathbf{z}^T] & \mathbb{E}[\mathbf{f}\mathbf{f}^T] \end{pmatrix}$$We already know $\mathbb{E}[\mathbf{z}\mathbf{z}^T] = \boldsymbol{\Sigma} = \boldsymbol{\Lambda}\boldsymbol{\Lambda}^T + \boldsymbol{\Psi}$ and $\mathbb{E}[\mathbf{f}\mathbf{f}^T] = \mathbf{I}_m$. For the cross terms:$$\mathbb{E}[\mathbf{z}\mathbf{f}^T] = \mathbb{E}[(\boldsymbol{\Lambda}\mathbf{f} + \boldsymbol{\epsilon})\mathbf{f}^T] = \boldsymbol{\Lambda}\mathbb{E}[\mathbf{f}\mathbf{f}^T] + \mathbb{E}[\boldsymbol{\epsilon}\mathbf{f}^T] = \boldsymbol{\Lambda}(\mathbf{I}_m) + \mathbf{0} = \boldsymbol{\Lambda}$$$$\mathbb{E}[\mathbf{f}\mathbf{z}^T] = (\mathbb{E}[\mathbf{z}\mathbf{f}^T])^T = \boldsymbol{\Lambda}^T$$Substituting these components yields the joint partitioned matrix:$$\mathbf{Cov} \begin{pmatrix} \mathbf{z} \\ \mathbf{f} \end{pmatrix} = \begin{pmatrix} \boldsymbol{\Lambda}\boldsymbol{\Lambda}^T + \boldsymbol{\Psi} & \boldsymbol{\Lambda} \\ \boldsymbol{\Lambda}^T & \mathbf{I}_m \end{pmatrix}$$Under Multivariate Normal assumptions, the conditional expectation identity for partitioned systems states that given a system matrix $\boldsymbol{\Sigma}_{\text{joint}} = \begin{pmatrix} \boldsymbol{\Sigma}_{11} & \boldsymbol{\Sigma}_{12} \\ \boldsymbol{\Sigma}_{21} & \boldsymbol{\Sigma}_{22} \end{pmatrix}$, the conditional expectation is:$$\mathbb{E}[\mathbf{X}_2 \mid \mathbf{X}_1] = \boldsymbol{\mu}_2 + \boldsymbol{\Sigma}_{21}\boldsymbol{\Sigma}_{11}^{-1}(\mathbf{X}_1 - \boldsymbol{\mu}_1)$$Mapping this directly to our factor score extraction system ($\mathbf{X}_2 = \mathbf{f}$, $\mathbf{X}_1 = \mathbf{z}$):$$\boldsymbol{\Sigma}_{21} = \boldsymbol{\Lambda}^T, \quad \boldsymbol{\Sigma}_{11} = \boldsymbol{\Sigma} = (\boldsymbol{\Lambda}\boldsymbol{\Lambda}^T + \boldsymbol{\Psi})$$$$\mathbb{E}[\mathbf{f} \mid \mathbf{z}] = \mathbf{0} + \boldsymbol{\Lambda}^T (\boldsymbol{\Lambda}\boldsymbol{\Lambda}^T + \boldsymbol{\Psi})^{-1}(\mathbf{z} - \mathbf{0})$$$$\hat{\mathbf{f}} = \boldsymbol{\Lambda}^T (\boldsymbol{\Lambda}\boldsymbol{\Lambda}^T + \boldsymbol{\Psi})^{-1}\mathbf{z}$$This completes the derivation of Thomson's Minimum Mean Squared Error estimator equation.

# **Part B: Parametric Factor Partition Engine & UI Architecture**

This section provides the implementation of the comprehensive engineering pipeline code. It includes preprocessing guardrails, a factor analysis engine with Varimax rotation, score calculation via Thomson's method, and a Plotly dashboard display.

```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import plotly.subplots as sp
import plotly.graph_objects as go

def varimax_rotation(Lambda, max_iter=1000, tolerance=1e-6):
    """
    Performs an orthogonal Varimax rotation on the factor loading matrix Lambda.
    """
    p, k = Lambda.shape
    R = np.eye(k)
    d = 0
    for i in range(max_iter):
        old_d = d
        Lambda_rot = Lambda @ R
        # Vectorized varimax step calculation
        transformed = Lambda_rot**3 - (1.0 / p) * Lambda_rot @ np.diag(np.sum(Lambda_rot**2, axis=0))
        u, s, vh = np.linalg.svd(Lambda.T @ transformed)
        R = u @ vh
        d = np.sum(s)
        if old_d != 0 and (d - old_d) / old_d < tolerance:
            break
    return Lambda @ R, R

def fit_parametric_fa_engine(df: pd.DataFrame, n_factors: int = 2) -> dict:
    """
    Validates, standardizes, and decomposes raw sensor matrices into common structural loads
    and unique noise floors, computing factor scores via Thomson's method.
    """
    N, p = df.shape
    numeric_df = df.select_dtypes(include=[np.number])
    feature_names = numeric_df.columns.tolist()
    X_raw = numeric_df.to_numpy()

    # --- Pre-processing Guardrails ---
    # Guardrail 1: Zero-variance catch
    stds = np.std(X_raw, axis=0, ddof=1)
    stds_capped = np.where(stds < 1e-9, 1e-5, stds)

    # Z-Score Layer
    means = np.mean(X_raw, axis=0)
    Z = (X_raw - means) / stds_capped

    # Guardrail 2: Dimensionality Constraints
    if N <= p:
        raise ValueError(f"Operational Exception: Snapshots ({N}) must exceed Feature count ({p}).")
    if n_factors >= p:
        raise ValueError(f"Constraint Exception: Requested factors ({n_factors}) must be < Feature count ({p}).")

    # --- Information Decomposition Layer via Principal Factor Method ---
    R_corr = np.corrcoef(Z.T)
    # Principal Factor Analysis iteratively estimates communalities via largest correlations
    h2_est = np.max(R_corr - np.eye(p), axis=0)

    # Single optimization iteration loop for stable loading separation
    for _ in range(3):
        R_reduced = R_corr.copy()
        np.fill_diagonal(R_reduced, h2_est)
        evals, evecs = np.linalg.eigh(R_reduced)
        idx = np.argsort(evals)[::-1]
        evals = evals[idx]
        evecs = evecs[idx]

        # Calculate unrotated loadings
        Lambda_raw = evecs[:, :n_factors] * np.sqrt(np.maximum(0, evals[:n_factors]))
        h2_est = np.sum(Lambda_raw**2, axis=1)

    # Apply orthogonal Varimax rotation
    Lambda_rotated, rot_matrix = varimax_rotation(Lambda_raw)

    # Calculate Uniqueness
    communalities = np.sum(Lambda_rotated**2, axis=1)
    uniquenesses = np.maximum(1e-5, 1.0 - communalities) # Protect floor bound

    # --- Latent Score Projection via Thomson's Method ---
    Psi_inv = np.diag(1.0 / uniquenesses)
    Sigma_inv = Psi_inv - Psi_inv @ Lambda_rotated @ np.linalg.inv(np.eye(n_factors) + Lambda_rotated.T @ Psi_inv @ Lambda_rotated) @ Lambda_rotated.T @ Psi_inv
    W_thomson = Lambda_rotated.T @ Sigma_inv

    factor_scores = Z @ W_thomson.T # Transformed vector mapping state array
    empirical_vars = np.var(factor_scores, axis=0, ddof=1)

    # Telemetry logging console confirmation
    print("--- PARAMETRIC MODEL TELEMETRY PROFILE ---")
    for f_idx, f_var in enumerate(empirical_vars):
        print(f" Latent Dimension Factor {f_idx + 1} Reconstructed Empirical Variance: {f_var:.5f}")

    return {
        "Features": feature_names,
        "Loadings": Lambda_rotated,
        "Communalities": communalities,
        "Uniquenesses": uniquenesses,
        "Factor_Variance": empirical_vars,
        "Scores": factor_scores
    }

def generate_dashboard_canvas(fa_results: dict):
    """
    Renders a unified production-grade 2x2 diagnostic canvas using Plotly subplots.
    """
    features = fa_results["Features"]
    loadings = fa_results["Loadings"]
    communalities = fa_results["Communalities"]
    uniquenesses = fa_results["Uniquenesses"]
    f_vars = fa_results["Factor_Variance"]

    n_fac = loadings.shape[1]
    factor_labels = [f"Factor_{i+1}" for i in range(n_fac)]

    fig = sp.make_subplots(
        rows=2, cols=2,
        horizontal_spacing=0.24,
        vertical_spacing=0.28,
        subplot_titles=(
            "Structural Loadings Matrix Heatmap",
            "Variance Partitioning Profile",
            "Sensor Uniqueness Line Profile",
            "Latent Factor Variance Distribution"
        )
    )

    # Subplot (1,1): Structural Loadings Heatmap
    heatmap = go.Heatmap(
        z=np.abs(loadings),
        x=factor_labels,
        y=features,
        colorscale='YlOrRd',
        colorbar=dict(
            x=-0.15, len=0.38, y=0.78,
            yanchor="middle", xanchor="right",
            title="Weight"
        ),
        showlegend=False
    )
    fig.add_trace(heatmap, row=1, col=1)

    # Subplot (1,2): Variance Partitioning Profile
    fig.add_trace(go.Bar(
        y=features, x=communalities * 100,
        name="Shared Communality (h²)", orientation='h',
        marker=dict(color='#1f77b4'),
        legendgroup="var_part"
    ), row=1, col=2)

    fig.add_trace(go.Bar(
        y=features, x=uniquenesses * 100,
        name="Isolated Uniqueness (φ²)", orientation='h',
        marker=dict(color='#ff7f0e'),
        legendgroup="var_part"
    ), row=1, col=2)

    # Subplot (2,1): Sensor Uniqueness Line Profile
    fig.add_trace(go.Scatter(
        x=features, y=uniquenesses,
        mode='lines+markers',
        name='Uniqueness Noise Floor',
        line=dict(color='#d62728', width=2, dash='dashdot'),
        marker=dict(symbol='x', size=8)
    ), row=2, col=1)

    # Subplot (2,2): Latent Factor Variance Distribution
    fig.add_trace(go.Bar(
        x=factor_labels, y=f_vars,
        name='Factor Variance Energy',
        marker=dict(color='#2ca02c', line=dict(color='black', width=0.5))
    ), row=2, col=2)

    # Layout Engineering Configuration updates
    fig.update_layout(
        template='plotly_white',
        width=1250, height=750,
        margin=dict(t=150, b=60, l=140, r=80),
        barmode='stack',
        legend=dict(orientation="h", x=0.5, y=1.12, xanchor="center")
    )

    fig.update_xaxes(range=[0, 100], row=1, col=2)
    fig.update_xaxes(tickangle=25, row=2, col=1)

    fig.show()

# --- Section 3 Block Execution Driver ---
# Instantiate pipeline data asset frame parameters from Section 4 requirement definition
np.random.seed(42)
n_samples_verify = 2500
f1_v = np.random.normal(0, 1, n_samples_verify)
f2_v = np.random.normal(0, 1, n_samples_verify)

s1_v = 0.85 * f1_v + 0.10 * f2_v + np.random.normal(0, 0.3, n_samples_verify)
s2_v = 0.80 * f1_v + 0.15 * f2_v + np.random.normal(0, 0.35, n_samples_verify)
s3_v = 0.12 * f1_v + 0.90 * f2_v + np.random.normal(0, 0.25, n_samples_verify)
s4_v = 0.02 * f1_v + 0.05 * f2_v + np.random.normal(0, 1.40, n_samples_verify)

df_asset_verify = pd.DataFrame(
    data=np.vstack([s1_v, s2_v, s3_v, s4_v]).T,
    columns=['Sensor_1', 'Sensor_2', 'Sensor_3', 'Sensor_4']
)

# Run complete engine
fa_output = fit_parametric_fa_engine(df_asset_verify, n_factors=2)
# To display the visualization plot inside an active runtime environment environment,
generate_dashboard_canvas(fa_output)
```

## **Part B: Comparative Diagnostic Analysis**

**Question 1: The Total Variance IllusionPhysical Nature of Sensor_4:**

A uniqueness value ($\psi^2$) exceeding $99\%$ confirms that Sensor_4 contains almost no common structural information. Physically, this means Sensor_4 is completely decoupled from the asset's true mechanical behaviors ($f_1$ and $f_2$) and is recording only localized white noise or a severe electrical/instrumentation failure.

How Sensor_4 Tricks the PCA Engine:

PCA constructs components by maximizing total global variance ($\text{Tr}(\boldsymbol{\Sigma})$) without differentiating between shared structural variance and localized random noise. Because the noise injection gives Sensor_4 an exceptionally large variance magnitude, PCA is forced to align its top principal components with this noise source to maximize variance capture. This inflates the top eigenvalues, making a noisy channel look like a dominant structural component.Automated Plant Framework Risks: If an engineer relies solely on PCA, the automated diagnostic layer will establish normal operational limits that are heavily influenced by sensor noise rather than actual structural behavior. Consequently, if Sensor_4's noise floor shifts or calms down, the system will trigger false alarms. More dangerously, true structural damage (drifts in Sensors 1, 2, or 3) will be obscured by the inflated noise thresholds, leading to missed detections of catastrophic structural failures.

**Question 2: Decoupling Structural Loading via RotationVarimax Rotation Mechanics vs. PCA Hierarchy:**

PCA enforces an orthogonal coordinate system optimized purely for variance maximization, which often spreads sensor contributions across multiple axes (cross-loading). Varimax rotation relaxes this strict variance hierarchy. It takes the retained subspace and rotates the coordinate axes to drive the factor loadings toward either 1 or 0. This isolates distinct groups of mutually correlated sensors on specific factors, creating a clean, unmixed layout.Operational Troubleshooting Benefit: Under unrotated PCA, a failure in a specific structural mode causes coordinates across multiple principal components to drift, making it difficult for an operator to isolate the root cause. The Varimax-rotated FA matrix maps specific physical processes to single factors. For example, if an operator sees an anomaly spike isolated to Factor 1, they know it specifically involves the process monitored by Sensors 1 and 2, which speeds up troubleshooting and root-cause analysis.

**Question 3: Determining Subspace Truncation ($k$) Using $T^2$ and SPE ProfilesSPE Statistic Curve Behavior:**

As the subspace cutoff moves from $k=0 \to 1$, the mean SPE drops from the total system variance down to $1.9486$. When moving from $k=1 \to 2$, SPE drops sharply to $0.8463$. However, transitioning from $k=2 \to 3$ or $k=3 \to 4$ produces a minimal reduction, with the curve flattening out into a clear horizontal line.Identifying Hidden Physical Dimensionality: The sharp drop followed by a flat "elbow" at $k=2$ identifies two as the true hidden physical dimensionality of the asset. The initial steep decline shows that the first two components capture the true, low-dimensional structural driving forces ($f_1$ and $f_2$). The flat elbow indicates that adding components beyond $k=2$ captures only uncorrelated white noise, which yields no meaningful improvement in reconstruction error.Consequence of Incorrect Truncation ($k=3$): Choosing a truncation cutoff of $k=3$ forces the high-magnitude localized noise of Sensor_4 directly into the "clean" principal subspace. This corrupts the baseline model by treating random instrument noise as a stable, predictable structural behavior, which undermines the model's diagnostic reliability.

**Which strategy is more robust against a single sensor losing calibration or experiencing an electrical short? Justify your choice using the Sensor Uniqueness Noise Floor ($\psi^2$) metrics.**

The Factor Analysis Strategy is significantly more robust against a single sensor losing calibration or experiencing an electrical short circuit.Justification: When a sensor fails, it injects a massive burst of uncorrelated variance into the data stream. As shown in Table 2, the Factor Analysis engine identifies this localized variance and allocates it almost entirely to the sensor's uniqueness metric ($\psi_4^2 = 99.81\%$). Because this localized noise is isolated within the diagonal noise matrix $\boldsymbol{\Psi}$, it is filtered out and prevented from entering the core factor loading matrix $\boldsymbol{\Lambda}$.Consequently, the estimated factor scores $\hat{\mathbf{f}}$ remain stable and continue to accurately reflect the true underlying state of the asset. In contrast, the PCA strategy would absorb this noise into its principal components, causing a spike in both Hotelling's $T^2$ and SPE, which would trigger a false alarm for a structural failure when only a sensor failed.
